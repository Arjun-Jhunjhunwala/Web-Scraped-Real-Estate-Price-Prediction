```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Raw Property.csv')
```

```python
df
```

```python
print(df.columns)
```

```python
nan_counts = df.isna().sum()
nan_counts
```

```python
df=df.dropna(how='all')
df.reset_index(drop=True, inplace=True)
df
```

```python
nan_counts = df.isna().sum()
nan_counts
```

```python
df['Built-up Area'] = df['Built-up Area'].str.split(' ').str[0].astype(int)
df['Bedroom'] = df['Bedroom'].str.split(' ').str[0].astype(float)
df['Price Unit'] = df['Price'].str.split(' ').str[1]
for i in range (0,len(df)):
    if df['Price Unit'][i] == 'Cr':
        df['Price'][i] = float(df['Price'][i].split(' ')[0]) * 10000000
    elif df['Price Unit'][i] == 'L' or df['Price Unit'][i] == 'Lacs' or df['Price Unit'][i] == 'Lac' or df['Price Unit'][i] == 'L(negotiable)':
        df['Price'][i] = float(df['Price'][i].split(' ')[0]) * 100000
    else:
      df['Price'][i] = float(df['Price'][i].replace(',',''))
df['Property_Age'] = df['Property_Age'].dropna().apply(lambda x: float(x.split()[0]))
df['Property_Age'].fillna(df['Property_Age'].median(), inplace=True)
df['Property_Name'] = df['Property_Name'].fillna('Unnamed Property')
df['Furnish'] = df['Furnish'].fillna('Unfurnished')
df
```

```python
df = df.drop_duplicates(ignore_index=True)
df
```
